```julia
using Plots
using QuadGK
global θ=1.0
global μ=100.0
global σ=1.0
global t0=0.0
global dt=0.01
global T=1000.0
```

```julia
#积分表达比级数表达快
function ET(p)
    α=sqrt(2*θ)/σ*p
    return sqrt(π/2)/θ*quadgk(t->(1+erf(t/sqrt(2)))*exp(t^2/2),0,α)[1]
end
p=-1:0.01:1
plot(p,ET.(p))
```

```julia
function yield(p)
    return p/ET(p)
end
plot(p,yield.(p))
```
