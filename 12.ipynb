```julia
#小N的怎么估μ̂？
#不要认为X1随机
#认为X1给定
#看下面，好起来了
#所以窗口的起始点有本质作用
```

```julia
using DifferentialEquations
using Plots
using Statistics
global θ=1.0
global σ=1.0
global μ=100.0
global dt=0.1
global T0=0.0
global T=1.0
X=OrnsteinUhlenbeckProcess(θ,μ,σ,T0,μ+10)
X.dt=dt
calculate_step!(X,dt,nothing,nothing)
for t in dt:dt:T
    accept_step!(X,dt,nothing,nothing)
end
plot(X.t,X.u,label="OU")
EX=mean(@view X.u[1:end-1])
plot!(X.t,fill(EX,length(X.u)),label="mean")
MLE = EX + (X.u[end]-X.u[1])/(θ*T) 
plot!(X.t,fill(MLE,length(X.u)),label="MLE")
```

```julia
global θ=1.0
global σ=1.0
global μ=100.0
global dt=0.1
global T0=0.0
global T=50.0
X=OrnsteinUhlenbeckProcess(θ,μ,σ,T0,μ+20)
X.dt=dt
calculate_step!(X,dt,nothing,nothing)
for t in dt:dt:T
    accept_step!(X,dt,nothing,nothing)
end
plot(X.t,X.u,label="OU",linewidth=0.4)
function MVUE(u)
    return mean(@view u[1:end-1]) + (u[end]-u[1])/(θ*dt*(length(u)-1)) 
end
function moving_MVUE(u,n)
    n=Int(round(n))
    return vcat( fill(NaN,n-1)  , [MVUE(@view u[i-n+1:i]) for i in n:length(u)] )
end
function moving_average(u,n)
    n=Int(round(n))
    return vcat( fill(NaN,n-1)  , [mean(@view u[i-n+1:i]) for i in n:length(u)] )
end
plot!(X.t,moving_average(X.u,1.0/dt),label="average",linewidth=1.0)
plot!(X.t,moving_MVUE(X.u,1.0/dt),label="MVUE",linewidth=1.0)
```

```julia
#可以看到，相比average，MLE(MVUE)更加重视价格的变化方向
#当距离均衡价格很远的时候，MVUE好到爆炸
#但是在均衡附近，MVUE比average起伏大
#但这是好处而不是坏处，至少不一定是坏的
#人认为的光滑不一定就好
```
