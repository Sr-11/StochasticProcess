```julia
# 在这里，你会看到，为什么数值模拟必须要数学证明
# 离散化会造成本质的问题
using DifferentialEquations
using Plots
using KernelDensity
using Statistics
using CUDA
using BenchmarkTools
println("nthreads=",Threads.nthreads())
#println("nthreads=",ENV["JULIA_NUM_THREADS"])
#CUDA.versioninfo()
```

```julia
N=100
x=Array{Float64}(undef,N)
dt = 0.0001
T=1.0
steps=floor(Int,T/dt)
u = nothing; p = nothing
function multi_threading()
    Threads.@threads for i = 1:N
        W = WienerProcess(0.0,0.0)
        W.dt = dt
        calculate_step!(W,dt,u,p)
        for i in 1:steps
          accept_step!(W,dt,u,p)
        end
        @inbounds x[i]=W.u[steps+1]
    end
end 
@time multi_threading()
U = kde(x)
p=plot(U.x,U.density)
plot!(p,U.x,1/sqrt(2π)*exp.(-(U.x).^2/2))
println("μ=",mean(x))
println("σ=",std(x))
display(p)
```

```julia
dt = 0.0001
T=1.0
steps=floor(Int,T/dt)
ub=0.02
lb=-0.01
deposit=0.0
u=nothing;p=nothing
W0=1.0
function one_epoch()
    W = WienerProcess(0.0,W0)
    W.dt = dt
    global deposit-=W0
    calculate_step!(W,dt,u,p)
    fig=plot()
    for t in dt:dt:T
        accept_step!(W,dt,u,p)
        if W.u[end]>=ub
            global deposit+=W.u[end]
            break
        elseif W.u[end]<=lb
            global deposit+=W.u[end]
            break
        end
    end
    return plot(W.t,W.u)
end
function simulation()
    Y=Array{Float64}(undef,1000)
    for i in 1:1000
        one_epoch()
        Y[i]=deposit
    end
    return plot(1:1000,Y)
end
fig=simulation()
xlabel!(fig,"epochs")
ylabel!(fig,"deposit")
title!(fig,"Deposit-Epoch")
```

```julia
# GBM 几何布朗运动
deposit=0.0
σ=0.01 # the percentage volatility
μ=σ^2/2 # the percentage drift
W0=100.0
W=GeometricBrownianMotionProcess(μ,σ,0.0,W0)
dt=0.001
W.dt=dt
u=nothing;p=nothing
lb=W0-1;lb=-Inf
ub=W0+2;ub=+Inf
T=1.0
calculate_step!(W,dt,u,p)
for t in dt:dt:T
    accept_step!(W,dt,u,p)
    if W.u[end]>=ub
        global deposit+=ub
        break
    elseif W.u[end]<=lb
        global deposit+=lb
        break
    end
end
fig=plot(W.t,W.u,xlabel="time",ylabel="quote",title="quote-time")
```

```julia
# GBM 几何布朗运动
deposit=0.0
σ=0.01 # the percentage volatility
μ=σ^2/2 # the percentage drift
W0=100.0
dt=0.001
u=nothing;p=nothing
ub=0.2
lb=-0.1
T=1.0
function one_epoch()
    W = GeometricBrownianMotionProcess(μ,σ,0.0,W0); W.dt = dt
    calculate_step!(W,dt,u,p)
    fig=plot()
    for t in dt:dt:T
        accept_step!(W,dt,u,p)
        if W.u[end]>=W0+ub
            global deposit+=ub
            break
        elseif W.u[end]<=W0+lb
            global deposit+=lb
            break
        end
    end
    return plot(W.t,W.u)
end
function simulation()
    Y=Array{Float64}(undef,1000)
    for i in 1:1000
        one_epoch()
        Y[i]=deposit
    end
    return plot(1:1000,Y)
end
fig=simulation()
xlabel!(fig,"epochs")
ylabel!(fig,"deposit")
title!(fig,"deposit-epochs")
```

```julia
x=CUDA.fill(1.0f0, N)
function gpu(x)
    N=10
    dt = 0.1
    t=1.0
    T=floor(Int,t/dt)
    u = nothing; p = nothing
    index = threadIdx().x 
    stride = blockDim().x
    for i=index:stride:length(x)
        @inbounds x[i]+=dt
    end
    return nothing
end 
@cuda threads=256 gpu(x)
```
