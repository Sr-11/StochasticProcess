```julia
# 现在，可以有策略了吗
# 统计套利的基石：Ornstein–Uhlenbeck Process
# https://zhuanlan.zhihu.com/p/521243637
# OU还是马尔可夫的
# 但是没有平移不变性了
# 只看其中我手里其中一股的行为
# 假设W0=μ=0
# 假设在p处卖出的概率是f(p)
# 卖出的收益是p
# 假设我现在留着，未来的收益的期望是E(p)
# 一定有f(p)=0或f(p)=1或p=E(p)
# f(p)=1相当于吸收
# 等效于在lb和ub处吸收的混合型随机变量,lb,ub是由我来定
# 现在有∀p,p=q而且是稳定的
# 把q算出来
# 过程叫做 Wₜ
# 首达时 Tₐ=min{t|Wₜ=a}
# 最简单的构造：>0直接卖，<0留着
# 直觉上,lb应该是-∞不存在,因为总会回归回来
# 记ub=m
```

```julia
Θ=1.0
μ=100.0
σ=1.0
t0=0.0
X0=100.0
dt=0.001
T=100.0
u=nothing;p=nothing
deposit=0.0
function one_epoch()
    global deposit-=μ
    X=OrnsteinUhlenbeckProcess(Θ,μ,σ,t0,X0)
    X.dt = dt
    calculate_step!(X,dt,u,p)
    for t in dt:dt:T
        accept_step!(X,dt,u,p)
        if X.u[end]>μ
            deposit+=X.u[end]
            return
        end
    end
    deposit+=X.u[end]
end
function simulation()
    Y=Array{Float64}(undef,1000)
    for i in 1:1000
        one_epoch()
        Y[i]=deposit
    end
    return plot(1:1000,Y)
end
simulation()
```
