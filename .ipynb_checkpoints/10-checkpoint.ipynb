```julia
using DifferentialEquations
using Plots
using Statistics
global θ=0.1
global σ=4.0
global μ=100.0
global dt=0.01
global T0=0.0
global T=100.0
X=OrnsteinUhlenbeckProcess(θ,μ,σ,T0,μ)
X.dt=dt
calculate_step!(X,dt,nothing,nothing)
for t in dt:dt:T
    accept_step!(X,dt,nothing,nothing)
end
plot(X.t,X.u,label="OU",linewidth=0.4)
EX=mean(X.u)
plot!(X.t,fill(EX,length(X.u)),label="mean")
function moving_average(u,n)
    n=Int(round(n))
    return vcat( fill(NaN,n-1)  , [mean(@view u[i-n+1:i]) for i in n:length(u)] )
end
function accumulate_average(u)
    return accumulate(+,u)./(1:length(u))
end
plot!(X.t,moving_average(X.u,5.0/dt),label="5.0",linewidth=1.2)
plot!(X.t,moving_average(X.u,10.0/dt),label="10.0",linewidth=1.2)
#plot!(X.t,moving_average(X.u,20.0/dt),label="20.0",linewidth=1.2)
plot!(X.t,accumulate_average(X.u),label="accumulate",linewidth=1.4)
#移动均线法 trend line 提供了对于 μ 的最佳估计
```
