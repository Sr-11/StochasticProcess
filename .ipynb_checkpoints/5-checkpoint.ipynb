```julia
using SpecialFunctions
using DifferentialEquations
using Plots
using QuadGK
using Statistics
using BenchmarkTools
global θ=1.0
global μ=100.0
global σ=1.0
global t0=0.0
global dt=0.01
global T=1000.0
```

```julia
function FPT(a,b) #首达时间
    if a==b
        return 0.0
    end 
    X=OrnsteinUhlenbeckProcess(Θ,μ,σ,t0,μ+a)
    X.dt=dt
    calculate_step!(X,dt,nothing,nothing)
    for t in dt:dt:T
        accept_step!(X,dt,nothing,nothing)
        if a<b && X.curW>=μ+b
            return X.curt
        end
        if a>b && X.curW<=μ+b
            return X.curt
        end
    end
    return Inf
end
function ET_experiment(a,b)
    N=100
    S=Array{Float64}(undef,N)
    for i in 1:N
        S[i]=FPT(a,b)
    end
    r=findall(x->x!=Inf,S)
    if length(r)!=length(S)
        println("P(reached)=",length(r)/length(S))
        println("T is too small when a=$a b=$b")
        error()
    end
    return mean(S[r])
end
```

```julia
function ET_series(a,b)
    if a>b
        a=-a;b=-b
    end
    α=sqrt(2*θ)/σ*a
    β=sqrt(2*Θ)/σ*b
    N=big(50)
    A=1/(2*Θ)*sum([(sqrt(2)*α)^n/factorial(n)*gamma(n/2) for n in 1:N])
    B=1/(2*Θ)*sum([(sqrt(2)*β)^n/factorial(n)*gamma(n/2) for n in 1:N])
    return B-A
end
```

```julia
function ET_integral(a,b)
    if a>b
        a=-a;b=-b
    end
    α=sqrt(2*θ)/σ*a
    A=sqrt(π/2)/θ*quadgk(t->(1+erf(t/sqrt(2)))*exp(t^2/2),0,α)[1]
    β=sqrt(2*θ)/σ*b
    B=sqrt(π/2)/θ*quadgk(t->(1+erf(t/sqrt(2)))*exp(t^2/2),0,β)[1]
    return B-A
end
```

```julia
function pdf0(a,t)
    if t==0
        return 0.0
    end
    α=a/σ
    term1=α/sqrt(2π)
    term2=(θ/sinh(θ*t))^(1.5)
    term3=exp(-θ*α^2*exp(-θ*t)/(2*sinh(θ*t))+θ*t/2)
    #问题就出在abs不是解析的？
    return term1*term2*term3
end
function ET0(a)
    if a==0
        return 0
    else
        f(t)=t*pdf0(a,t)
        return quadgk(f,0,1000)[1]
    end
end
function ET_trick(a,b)
    #我们已经知道 ET(a,b) = f(a)-f(b)(a<b), f(-a)-f(-b)(a>b)
    #但是f本身并不是正负对称的
    #不妨f(0)=0
    #我们还知道 ET0(a)=ET(a,0)=f(a)(a<0), f(-a)(a>0)
    #所以确定下来f(<0)=ET0()
    #但是f(>0)永远不知道
    #看图就知道，输出的是折回来的部分
    #我们能不能用数值实现解析延拓？？？
    #！！！！！数值的解析延拓！！！！！
    #但是我们现在的表达式不就是解析的？
    #难道说无穷积分会破坏解析性质？
    #不不不，我们只知道负实半轴的信息
    #不足够唯一确定
    #那救不回来了？
    #不知道
    return ET0(a)-ET0(b)
end
```

```julia
global Θ=1.0
global σ=1.0
x=0:0.1:1
y_experiment=ET_experiment.(0,x)
y_series=ET_series.(0,x)
y_integral=ET_integral.(0,x)
y_trick=ET_trick.(0,x)
fig=plot(x,y_experiment,xlabel="0→x",ylabel="𝔼T(0→x)",title="𝔼T(0→x)",label="experiment")
plot!(fig,x,y_series,label="series")
plot!(fig,x,y_integral,label="integral")
plot!(fig,x,y_trick,label="trick")
#看不见另外一条是因为重叠住了
```
