```julia
# 在p<0买入
# UO过程收敛在一个正太，所以p可以足够小，如果不限制时间
# 还是效率最大化
# 亏钱速率最小化
# min |p|*ET(0->p)
# max p*ET(0->p) 但p<0
using Plots
using QuadGK
using SpecialFunctions
global θ=1.0
global σ=1.0
```

```julia
function ET(p)
    b=-p#(a=0,a<b)
    β=sqrt(2*θ)/σ*b
    return sqrt(π/2)/θ*quadgk(t->(1+erf(t/sqrt(2)))*exp(t^2/2),0,β)[1]
end
p=-1:0.01:1
plot(p,p.*ET.(p),xlabel="price",ylabel="profit")
```

```julia
#其实没有什么关系，一旦<0就买就行了，必不亏
#因为价格总是倾向于上升的，一定要买
#现在我们得到了完整的策略
#一旦价格比平均价格小，狂买入
#一旦价格比平均价格大，狂卖出
#而且这样是最有效率的方法
#下面的问题是，怎么从历史数据把μ定出来
#但和我们的结论没有关系，对估出来的μ持续以上操作就行
#说明什么？说明我们不能从市场的完全随机性里赚钱
```
